```python
import pandas as pd
```

```python
df=pd.read_csv('Data.csv', encoding = "ISO-8859-1")
```

```python
df.head()
```

```python
train = df[df['Date'] < '20150101']
test = df[df['Date'] > '20141231']
```

```python
# Dropping the punctuations
data=train.iloc[:,2:27]
data.replace("[^a-zA-Z]"," ",regex=True, inplace=True)

# Renaming the column names for our convinence
list1= [i for i in range(25)]
new_Index=[str(i) for i in list1]
data.columns= new_Index
data.head(5)
```

```python
# Transforming the headlines to lower case
for index in new_Index:
    data[index]=data[index].str.lower()
data.head(1)
```

```python
' '.join(str(x) for x in data.iloc[1,0:25])
```

```python
headlines = []
for row in range(0,len(data.index)):
    headlines.append(' '.join(str(x) for x in data.iloc[row,0:25]))
```

```python
headlines[0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
## implementing BAG OF WORDS Algorithm
countvector=CountVectorizer(ngram_range=(2,2))
traindataset=countvector.fit_transform(headlines)
```

```python
# implement RandomForest Classifier
randomclassifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(traindataset,train['Label'])
```

```python
## Predict for the Test Dataset
test_transform= []
for row in range(0,len(test.index)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row,2:27]))
test_dataset = countvector.transform(test_transform)
predictions = randomclassifier.predict(test_dataset)
```

```python
## Import library to check accuracy
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score, precision_score, recall_score, f1_score
```

```python
matrix=confusion_matrix(test['Label'],predictions)
print(matrix)
score=accuracy_score(test['Label'],predictions)
print(score)
report=classification_report(test['Label'],predictions)
print(report)
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.imshow(matrix, interpolation='nearest', cmap=plt.cm.Blues)
plt.title("Confusion Matrix")
plt.colorbar()

# Adding labels to each row and column
plt.tick_params(axis='both', which='major', labelsize=12)
class_names = ['neg', 'pos']  # Modify these labels as needed
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names, rotation=45)
plt.yticks(tick_marks, class_names)

# Adding a grid
plt.grid(True, color='w', linestyle='-', linewidth=2)

# Adding labels and title
plt.xlabel('Predicted class')
plt.ylabel('True class')
plt.show()

# Calculating and printing other metrics
accuracy = accuracy_score(test['Label'], predictions)
precision = precision_score(test['Label'], predictions)
recall = recall_score(test['Label'], predictions)
f1 = f1_score(test['Label'], predictions)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```
